```python
import os, cv2, random
import numpy as np
import matplotlib.pyplot as plt
```

```python
DATA_DIR = '../gender_dataset/'
CLASSES = os.listdir(DATA_DIR)
k = 5
n = len(CLASSES)
```

```python
random.seed(19)
IMAGES = []

for CLASS in CLASSES:
    img_paths = os.listdir(os.path.join(DATA_DIR, CLASS))
    imgs = []
    for img in random.choices(img_paths, k=k):
        imgs.append(cv2.resize(cv2.cvtColor(cv2.imread(os.path.join(DATA_DIR, CLASS, img)), cv2.COLOR_BGR2RGB), (96,96)))
    IMAGES.append(imgs)
```

```python
fig, ax = plt.subplots(nrows=n, ncols=k, figsize=(20, 4*len(CLASSES)))
for i in range(n):
    for j in range(k):
        ax[i][j].imshow(IMAGES[i][j])
        if j == k//2:
            ax[i][j].set_title(f'CLASS: {CLASSES[i]}', fontsize=32)
```
